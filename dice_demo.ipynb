```python
 def srmeq(alpha,beta,gamma,IS):

        return (alpha* np.exp(-(beta/IS ** gamma)))
    
```

```python
 def srmeq2(alpha,beta,gamma,IS):

        return ((IS/beta)**2)*alpha/15
```

```python
plt.plot(np.arange(0,10,.1),srmeq(10,8,0.5,np.arange(0,10,.1)))
plt.plot(np.arange(0,10,.1),srmeq2(10,8,0.5,np.arange(0,10,.1)))
```
